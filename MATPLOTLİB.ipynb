```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3,4]
y = [1,4,9,16]
# plt.plot(x,y,color = "red",linewidth = 5)
# plt.plot(x,y,"-g",linewidth = 4)

plt.plot(x,y,"o--r")  ## marker, type of line, color
plt.axis([0,6,0,20])
plt.title("TÄ°TLE OF GRAPH")
plt.xlabel("title of x label")
plt.ylabel("title of y label")
```

```python
x = np.linspace(0,2,100)
# plt.plot(x,x,label = "linear",color = "red")
# plt.plot(x,x**2,label = "quadratic",color = "yellow")
plt.plot(x,x**3,label = "cubic",color = "green")

plt.xlabel("x label")
plt.ylabel("y label")
plt.legend()
plt.title("simple plot")
```

```python
x = np.linspace(0,2,100)
fig,axs = plt.subplots(3)

axs[0].plot(x,x,color = "red")
axs[0].set_title("linear")

axs[1].plot(x,x**2,color = "green")
axs[1].set_title("quadratic")

axs[2].plot(x,x**3,color="blue")
axs[2].set_title("cubic")

plt.tight_layout() ##  grafiklerin iÃ§ iÃ§e geÃ§mesini engelliyor

plt.show()
```

```python
x = np.linspace(0,2,100)
fig,axs = plt.subplots(2,2)

axs[0,0].plot(x,x,color ="red")
axs[0,1].plot(x,x**2,color = "blue")
axs[1,0].plot(x,x**3,color="green")
axs[1,1].plot(x,x**4,color="yellow")
```

```python
x = np.linspace(-10,9,20)
y = x**3
z = x**2

figure = plt.figure()

# axes_cube = figure.add_axes([0.1,0.1,0.9,0.9])
# axes_cube.plot(x,y,"b")

# axes_cube.set_xlabel("X LABEL")
# axes_cube.set_ylabel("Y LABEL")
# axes_cube.set_title("cube")

axes_square=figure.add_axes([0.15,0.6,0.25,0.25])  
axes_square.plot(x,z,"r")
axes_square.set_xlabel("X AXÄ°S")
axes_square.set_ylabel("Y axis")
axes_square.set_title("square")

plt.show()
```

```python
x=np.linspace(-10,9,20)
y=x**3
z=x**2

figure= plt.figure()

axes=figure.add_axes([0,0,1,1])

axes.plot(x,z,label="Square")
axes.plot(x,y,label="Cube")
axes.legend(loc=4) #ALT SAÄ KÃÅEYE GELDÄ° BU 4 YAZAN YERÄ° 1,2,3, GÄ°BÄ° DEÄÄ°ÅTÄ°REBLÄ°RÄ°Z 

plt.show()
```

```python
x=np.linspace(-10,9,20)
y=x**3
z=x**2

fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,8)) #SÄ°ZE BOYUT BÃYÃTÃYOR.

axes[0].plot(x,y)
axes[0].set_title("Cube")
axes[1].plot(x,z)
axes[1].set_title("Square")

plt.tight_layout() #BAÅLIKLAR BÄ°RBÄ°RÄ°NE GÄ°RMESÄ°N DÄ°YE
```

```python
yil=[2011,2012,2013,2014,2015]

oyuncu1=[8,10,12,7,9]
oyuncu2=[7,12,5,15,21]
oyuncu3=[18,20,22,25,19]

plt.stackplot(yil, oyuncu1, oyuncu2, oyuncu3, colors = ["y","r","b"])
plt.title("YÄ±llara gÃ¶re atÄ±lan goller")
plt.xlabel("yil")
plt.ylabel("gol sayisi")
# plt.legend()
plt.show()
```

```python
####  PÄ°E GRAPH 

gol_types = ["penaltÄ±","kaleye atÄ±lan Åut","serbest vuruÅ"]

goals = [12,35,7]
colors = ["y","r","b"]
plt.pie(goals,labels=gol_types,colors = colors, shadow=True, explode=(0.05,0.05,0.05),autopct="%1.1f%%")
plt.show()
```

```python
####  BAR GRAPH

plt.bar([0.25,1.25,2.25,3.25,4.25],[50,40,70,80,20],label="BMW",width=.5)
plt.bar([0.75,1.75,2.75,3.75,4.75],[80,20,20,50,60],label="Audi",width=.5)

plt.legend()
plt.xlabel("gÃ¼n")
plt.ylabel("mesafe (km)")
plt.title("araÃ§ bilgileri")

plt.show()
```
